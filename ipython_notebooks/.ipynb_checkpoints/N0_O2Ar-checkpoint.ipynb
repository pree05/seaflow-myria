```python
% matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seawater as sw

```

```python
fileURL = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-N0_TS_Cruise/data?format=csv'
result = pd.read_csv(fileURL)
print pd.unique(result['Cruise'])
```

```python
# select data for a particular cruise

cruise = 'Tokyo_2' # this seems to be the best cruise for this
data = result.loc[(result['Cruise']==cruise)]
data = data.sort(['day','file'],ascending=[True,True])
print(data.columns)
```

```python
# read in the O2Ar data

file = '/Users/archie/Desktop/Tokyo2_O2Ar.csv'
o2ar = pd.read_csv(file)
print(o2ar.columns)
```

```python
o2_lon = o2ar['LON'].values
o2_lat = o2ar['LAT'].values
o2_lon[o2_lon<0]= 360+o2_lon[o2_lon<0];
n0_lon = data['LON'].values
n0_lat = data['LAT'].values
n0_lon[n0_lon<0]= 360+n0_lon[n0_lon<0];

o2_ar = o2ar['O2Arbiosat'].values
o2_ncp = o2ar['NCPc'].values

n0 = np.float32(data['N0'].values)

fig1 = plt.figure(1, figsize=(15,7))

plt.plot(o2_lon,o2_ar,'.b',n0_lon,n0/np.max(n0)*20,'.r')
plt.show()
```
