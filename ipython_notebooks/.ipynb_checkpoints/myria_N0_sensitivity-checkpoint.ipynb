```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the data
# 3-D matrix with T, S and the biological variable, e.g. mean bead fluorescence
fileURL1 = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-richness_untrans_8bins/data?format=csv'
result1 = pd.read_csv(fileURL1)

fileURL2 = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-richness_untrans_16bins/data?format=csv'
result2 = pd.read_csv(fileURL2)

fileURL3 = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-richness_untrans_32bins/data?format=csv'
result3 = pd.read_csv(fileURL3)

print(result1.columns)
print(result2.columns)
print(result3.columns)
```

```python
# select data from just one cruise as an example

cruise = 'MBARI_1'

data1 = result1.loc[(result1['Cruise']==cruise)]
data1.sort(['Day','File_Id'],ascending=[True,True])
y1 = np.float32(data1['richness'].values) # richness with 8 bins
data2 = result2.loc[(result2['Cruise']==cruise)] # richness with 16 bins
data2.sort(['Day','File_Id'],ascending=[True,True])
y2 = np.float32(data2['richness'].values)
data3 = result3.loc[(result3['Cruise']==cruise)] # richness with 32 bins
data3.sort(['Day','File_Id'],ascending=[True,True])
y3 = np.float32(data3['richness'].values)
```

```python
# plot
fig1, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,15))
#plt.plot(range(0,500),y1[0:500]/np.max(y1), range(0,500),y2[0:500]/16**3,range(0,500),y3[0:500]/32**3)
ax[0].plot(y2,y3,'o')
ax[1].plot(y2,y1,'o')

plt.show()
```
