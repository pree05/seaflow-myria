```python
% matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seawater as sw

from mpl_toolkits.basemap import Basemap, cm
from netCDF4 import Dataset as NetCDFFile
```

```python
# import data directly from myria

fileURL = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-N0_TS_Cruise/data?format=csv'
result = pd.read_csv(fileURL)
print pd.unique(result['Cruise'])
```

```python
# select data for a particular cruise

cruise = 'Tokyo_1'
data = result.loc[(result['Cruise']==cruise)]
data = data.sort(['day','file'],ascending=[True,True])
```

```python
# plot the mean bead fsc_small value in space
# the data
x = (data['LON']).values # longitude
y = (data['LAT']).values # latitude

S = (data['S']).values
T = (data['T']).values

z = np.float32((data['N0']).values) # bead attribute

# convert all longitude to longitude east
x[x<0]= 360+x[x<0];

print(max(x))
```

```python
fig1 = plt.figure(1, figsize=(15,7))

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(x, y, s=75, c=z/max(z), alpha=0.5,marker='o',edgecolor='none')

plt.xlabel('longitude',fontsize=16)
plt.ylabel('latitude',fontsize=16)
cb = plt.colorbar(shrink=.92)
cb.set_label(label='N0',fontsize=16)

plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
# Now look at N0 versus the density gradient

# calculate density and gradient
density = sw.dens(S,T,0)

# convert the longitude back to E/W
#x_ew=x
#x_ew[x_ew>180]=x_ew[x_ew>180]-180;
# calculate the distance between points
distance = sw.dist(y, x)[0]
track = np.cumsum(distance)
track = np.insert(track, 0, 0)

# density gradient
grad_dens = (density[1:len(density)]-density[:len(density)-1])/distance

fig2 = plt.figure(2, figsize=(15,10))

plt.plot(track, T,'ob', track, z/np.max(z)*10,'or')
plt.axis([0,np.max(track),0,30])
plt.ylabel('temperature (blue), re-scaled N0 (red)',fontsize=16)
plt.xlabel('along track distance',fontsize=16)
plt.title(cruise,fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
fig3 = plt.figure(3, figsize=(15,10))

plt.plot(track[1:len(track)], abs(grad_dens)*10,'-b', track, z/np.max(z),'or')
plt.axis([0,np.max(track),0,1])
plt.ylabel('absolute density gradient (blue), relative N0 (red)',fontsize=16)
plt.xlabel('along track distance',fontsize=16)
plt.title(cruise,fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
