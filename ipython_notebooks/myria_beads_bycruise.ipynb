```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the data
# 3-D matrix with T, S and the biological variable, e.g. mean bead fluorescence
fileURL = 'https://rest.myria.cs.washington.edu:1776/dataset/user-armbrustlab/program-seaflow/relation-bead_stats_v4_bycruise_untrans/data?format=csv'
result = pd.read_csv(fileURL)
```

```python
data = result.sort('chl_avg')

# the data
x = result['Cruise'] # cruise

y1 = data['fsc_avg'].values # mean of bead attribute
z1 = data['fsc_sd'].values # stdev of bead attribute
y2 = data['chl_avg'].values # mean of bead attribute
z2 = data['chl_sd'].values # stdev of bead attribute
y3 = data['pe_avg'].values # mean of bead attribute
z3 = data['pe_sd'].values # stdev of bead attribute

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18,7))

ax[0].errorbar(range(0,len(x)),y1,yerr=z1,fmt='o')
#ax[0,0].xticks(range(0,len(x)),x, rotation=90)
#ax[0,0].xlabel('Cruise')
#ax[0,0].ylabel('Bead fsc_small')
ax[0].axis((-1,len(x)+1,25000,65000))

ax[1].errorbar(range(0,len(x)),y2,yerr=z2,fmt='o')
#ax[0,1].xticks(range(0,len(x)),x, rotation=90)
#ax[0,1].xlabel('Cruise')
#ax[0,1].ylabel('Bead chl_small')
ax[1].axis((-1,len(x)+1,15000,50000))

ax[2].errorbar(range(0,len(x)),y3,yerr=z3,fmt='o')
#ax[0,1].xticks(range(0,len(x)),x, rotation=90)
#ax[0,1].xlabel('Cruise')
#ax[0,1].ylabel('Bead chl_small')
ax[2].axis((-1,len(x)+1,35000,65000))

plt.show()
```
