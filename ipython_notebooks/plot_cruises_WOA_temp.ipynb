```python
% matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.mlab import griddata
import seawater as sw
```

```python
# import WOA temperature climatology for 2005-2012

#url = 'http://data.nodc.noaa.gov/woa/WOA13/DATA/temperature/csv/A5B2/1.00/woa13_A5B2_t00mn01.csv.gz'
#request = urllib2.Request(url)
#request.add_header('Accept-encoding', 'gzip')
#response = urllib2.urlopen(request)

#buf = StringIO( response.read())
#f = gzip.GzipFile(fileobj=buf)

f = '/Users/archie/Desktop/woa13_A5B2_t00mn01.csv'
data = pd.read_csv(f)

print(data.columns)
```

```python
lat = data[' LATITUDE'].values
lon = data[' LONGITUDE'].values
sst = data['T'].values
```

```python
# define grid.
xi = np.linspace(-180, 180, 300)
yi = np.linspace(-80,80,120)
# grid the data.
zi = griddata(lon,lat,sst, xi, yi, interp='linear')
# contour the gridded data, plotting dots at the nonuniform data points.
CS = plt.contour(xi, yi, zi, 15, linewidths=0.5, colors='k')
CS = plt.contourf(xi, yi, zi, 15, cmap=plt.cm.rainbow,
                  vmax=abs(zi).max(), vmin=-abs(zi).max())
plt.colorbar()  # draw colorbar
# plot data points.
plt.xlim(-180,180)
plt.ylim(-70,70)

plt.show()
```
